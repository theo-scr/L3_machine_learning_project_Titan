### 1.2.6 - Base d'apprentissage et base de test

```python
# Fonction permettant de séparer le DataFrame initial en deux ensembles : la base d'apprentissage 'train' et la 
# de test 'test'.

def split(df_init : pd.core.frame.DataFrame, proportion: float):
    """ Documentation
    Paramètre :
        * df_init : DataFrame a séparer en deux
        * proportion (float) : proportion des données totales alliuées à la base d'apprentissage 'train'  

    Sortie :
        * df1 (DataFrame) : Correspond à la base d'apprentissage 'train'
        * df2 (DataFrame) : Correspond à la base de test 'test'
    """
    
    # Dans un premier temps, on stocke dans n le nb de lignes dans le DataFrame initial
    # On en déduit nTrain le nb de lignes dans le DataFrame associé à l'apprentissage
    n = df_init.shape[0]
    nTrain = int(n*proportion)
    
    # Dans un deuxième temps on mélange le DataFrame initial
    df_init = df_init.sample(n)
    
    # Enfin on le scinde en deux
    df1 = pd.DataFrame(df_init.iloc[:nTrain])
    df2 = pd.DataFrame(df_init.iloc[nTrain:])
    
    return (df1,df2)
```

```python
# On applique la fonction à notre DataFrame 'df'. On choisit 65% pour 'train' et donc 35% pour 'test'.
(train, test) = split(df, 65/100)
```

```python
# Tout d'abord, on définit la fonction permettant de calculer le taux de prédiction (en %).
# Le taux de prédiction est également appelé "taux de classification" (ou encore "accuracy" en anglais )
# La formule est : (1/n) * somme de (1-|yi - f(xi)|) avec n le nombre total de prédiction, yi la classe réelle et
# f(xi) la classe prédite, d'où: 

def prediction(labels : pd.core.series.Series, pred : pd.core.series.Series):
    """ Documentation
    Paramètre :
        * labels (Series) : contient les classes "réelles" (= yi)
        * pred (Series) : contient les classes prédites par le modèle (= f(xi))

    Sortie :
        * taux_prediction (float) : taux de prédiction en pourcentage avec deux chiffres après la virgule. 
    """
    
    # On compte le nombre de bonnes prédictions
    nbJuste = sum(labels == pred)
    
    # On récupère le nombre total de prédictions qu'il fallait faire
    taille = len(pred)

    # Le taux correspond au nombre de bonnes prédictions divisées par le nombre total
    return round(nbJuste/taille*100, 2)
```

## 2.3 - Comparaison des méthodes par validation croisée
Pour chacune des méthodes et en utilisant les paramètres optimaux déterminés précédemment, réalisez une procédure de validation croisée pour calculer le taux de prédiction moyen et sa variance lorsque le nombre de folds est 5.

```python
# Fonction permettant, à partir du tableau de scores (=taux de prédiction), de calculer leur moyenne et leur variance 


def moyenne_variance (scores : np.ndarray):
    """ Documentation
    Paramètre :
        * scores : Tableau contenant les taux de prédiction pour chaque fold d'une validation croisées, obtenu
        grâce à la fonction 'cross_val_scores'.

    Sortie :
        (moy) : float -> moyenne des taux de prédiction
        (var) : float -> variance des taux de prédiction
    """
    
    moy = sum(scores)/len(scores)
    
    var = sum((scores - moy)**2) / len(scores)
    
    return (moy, var)
```

```python
# Ici nous ne cherchons pas à optimiser les paramètres, il n'est donc pas nécessaire de garder un ensemble 'test'.
# La validation croisée peut se faire sur l'ensemble des données. (Remarque: la validation croisée est aussi 
# utilisée lorsqu'on dispose de peu de données et qu'il est impossible de séparer le DataFrame en deux ou trois).
# Il faut donc seulement séparer la classe ('Y') du reste des variables.

X = df.iloc[:, 1:]
Y = df.iloc[:,0]
```

```python
# Fonction permettant de séparer le DataFrame initial en trois ensembles : la base d'apprentissage 'train', la 
# de test pour la recherche des paramètres optimaux 'test' et la base de validation du modèle final 'validation'.

def split2(df_init, proportion_train, proportion_test):
    """ Documentation
    Paramètre :
        * df_init : DataFrame à séparer en trois
        * proportion_train (float) : proportion des données totales alliuées à la base d'apprentissage 'train' 
        * proportion_test (float) : proportion des données totales alliuées à la base d'apprentissage 'test'  

    Sortie :
        * df1 (DataFrame) : Correspond à la base d'apprentissage 'train'
        * df2 (DataFrame) : Correspond à la base de test 'test'
        * df3 (DataFrame) : Correspond à la base de validation 'validation'
    """
    
    
    # Dans un premier temps, on stocke dans n le nb de lignes dans le DataFrame initial
    # On en déduit nTrain le nb de lignes dans le DataFrame associé à l'apprentissage et nTest le nb de lignes 
    # dans le DataFrame associé au test des paramètres.
    n = df_init.shape[0]
    nTrain = int(round(n*proportion_train))
    nTest = int(round(n*proportion_test))
    
    # Dans un deuxième temps on mélange le DataFrame initial
    df_init = df_init.sample(n)
    
    # Enfin on le scinde en trois
    df1 = pd.DataFrame(df_init.iloc[:nTrain])
    df2 = pd.DataFrame(df_init.iloc[nTrain:(nTrain+nTest)])
    df3 = pd.DataFrame(df_init.iloc[(nTrain+nTest):])
    
    return (df1,df2,df3)
```

# 3 - Partie Entropie
Le but ici est de caculer l'entropie et de gain d'entropie comme pour la construction d'un arbre de décision.
* Ecrire une fonction entropie qui prend une liste (ou un array numpy) de valeur bianires et qui calcule l'entropie. Appliquer cette fonction sur la classe.
* Ecrire une fonction gain d'entropie qui prend une liste pour la classe (en binaire) et une liste pour les valeurs correspondantes d'un autre attribut (binaire aussi) 
* Calculer le gain d'entropie pour l'attribut pour le sexe et toutes les autres variables binaires que vous avez crées. Quelle serait le meilleur attribut pour démarrer un arbre de décision ?

```python
# Fonction permettant le calcul de l'entropy dans le cas binaire (lorsqu'il n'y a que deux classes). 


def entropie(liste: pd.core.series.Series):
    """ Documentation
    Paramètre :
        * liste : Series contenant les valeurs de la classe: 0 pour 'n'a pas survécu' et 1 pour 'a survécu' (binaire)

    Sortie :
        (entropy) : float entre 0.0 et 1.0 -> valeur de l'entropy (=l'impureté de l'ensemble d'exemple / nombre de
        bits nécessaires pour coder la classe d'un élément tiré au hasard (théorie de l'information))
    """
    # Total
    total = liste.count()
    
    # Proportion de cas "positifs" (0) et proportion de cas "négatifs" (1) 
    proportion_0 = liste.value_counts()[0] / total
    proportion_1 = liste.value_counts()[1] /total
    
    # Formule de l'entropie
    return (- proportion_0 * math.log2(proportion_0) - proportion_1 * math.log2(proportion_1))
```

```python
# Fonction permettant de calculer le gain d'entropy suite à la sélection d'un attribut.
# Remarques: 
#     * Pour utiliser cette fonction, il faut que les variables soient bien-sûr binaires mais aussi que leur 
# valeur soient uniquement 0 ou 1 (il faut donc nettoyer un peu les données avant d'utiliser la fonction).
#     * Le fait de passer le DataFrame en paramètre permet l'utilisation de la fonction, également pour les étapes
# 2 et au delà (lorsqu'un attribut est selectionné, l'arbre se développe et des branches sont créées donc le 
# DataFrame varie selon les branches). 


def gain_entropie(liste_class : pd.core.series.Series,  dataframe : pd.core.frame.DataFrame, attribut : str):
    """ Documentation
    Paramètre :
        * liste_class : Series contenant les valeurs de la classe: 0 pour 'n'a pas survécu' et 1 pour 
        'a survécu' (binaire)
        * dataframe : DataFrame contenant les données
        * attribut : Nom de l'attribut qu'on souhaite sélectionner

    Sortie :
        (gain_entropy) : float entre 0.0 et 1.0 -> valeur du gain d'entropy (= quantification de la réduction 
        d'entropie après la sélection d'un attribut)
    """
    
    # On distingue les cas "positifs" des cas "negatifs" et on récupère leur classe
    liste_ex_0 = dataframe.loc[dataframe[attribut] == 0]['Class']
    liste_ex_1 = dataframe.loc[dataframe[attribut] == 1]['Class']
    
    # Calcul des différentes entropies
    entropie_ex_0 = entropie(liste_ex_0)
    entropie_ex_1 = entropie(liste_ex_1)
    entropie_class = entropie(liste_class)
    
    # Calcul du poids de chaque cas
    poids_ex_0 = len(liste_ex_0) / len(liste_class)
    poids_ex_1 = len(liste_ex_1) / len(liste_class)
    
    # Formule du gain d'entropie
    return (entropie_class - poids_ex_0 * entropie_ex_0 - poids_ex_1 * entropie_ex_1)    
```
